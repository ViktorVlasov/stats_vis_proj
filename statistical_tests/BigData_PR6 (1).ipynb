## Task 2

```python
for column in df.columns:
    missing = np.mean(df[column].isna() * 100)
    print(f'{column} : {round(missing, 1)}%')
```

```python
df = df.drop(['Cumulative_number_for_14_days_of_COVID-19_cases_per_100000', 'geoId'], axis=1)
```

```python
categorical_attributes = [
    'dateRep',
    'countriesAndTerritories',
    'countryterritoryCode',
    'continentExp'
]
```

```python
for column in list(df.columns):
    if column in categorical_attributes:
        df[column] = df[column].fillna('other')
    else:
        df[column] = df[column].fillna(np.median(df[column]))
```

```python
df_bmi = pd.read_csv('bmi.csv')
df_bmi
```

```python
df_northwest = df_bmi[df_bmi.region == 'northwest'].reset_index(drop=True)
df_southwest = df_bmi[df_bmi.region == 'southwest'].reset_index(drop=True)
```

### Критерий Шапиро-Уилка

```python
print(f'Север: {sts.shapiro(df_northwest.bmi)}\n Юг: {sts.shapiro(df_southwest.bmi)}')
```

## Критерий Бартлетта

```python
print(sts.bartlett(df_southwest.bmi, df_northwest.bmi))
```

p-уровень превышает 0.05 незначительно, десперсии скорее всего имеют отличия

### Критерий Стьюдента

```python
print(sts.ttest_ind(df_southwest.bmi, df_northwest.bmi))
```

p-значение намного ниже 0.05, следовательно нулевая гипотеза отвергается, выборки принадлежат к разным генеральным совокупностям и их средние значения различны.
